```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('PhaseOne-SDK.jpg')
image2 = image[:,:,::-1]
plt.imshow(image2)
sift = cv2.xfeatures2d.SIFT_create()
#sift = cv2.xfeatures2d.SIFT_create()
kps, dscs = sift.detectAndCompute(image2, mask=None)
ax = plt.gca()
for kp in kps:
    ax.add_artist(plt.Circle((kp.pt), kp.size/2, color='green', fill=True))
    
plt.show()
```
