```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.draw import circle_perimeter_aa,line,line_aa,rectangle,set_color,ellipse_perimeter

fig,ax = plt.subplots(ncols=1,nrows=1,figsize=(10,6))
img = np.zeros((512,512,3),dtype=np.uint8)


#roads:
rr,cc = line(511,100,200,200)
img[rr,cc,0] = 255
img[rr,cc,1] = 255
img[rr,cc,2] = 255

rr,cc = line(511,110,200,210)
img[rr,cc,0] = 255
img[rr,cc,1] = 255
img[rr,cc,2] = 255

rr,cc,val = line_aa(201,201,0,201)
img[rr,cc,0] = val*255
img[rr,cc,1] = val*255
img[rr,cc,2] = val*255

rr,cc = line(202,212,0,212)
img[rr,cc,0] = 255
img[rr,cc,1] = 255
img[rr,cc,2] = 255

rr,cc,val = line_aa(301,0,301,511)
img[rr,cc,0] = val*255
img[rr,cc,1] = val*255
img[rr,cc,2] = val*255


rr,cc = line(315,0,315,511)
img[rr,cc,0] = 255
img[rr,cc,1] = 255
img[rr,cc,2] = 255




#river:
rr,cc = ellipse_perimeter(30,30,100,150,40,img.shape)
img[rr,cc,2] = 150

rr,cc = ellipse_perimeter(30,30,120,170,40,img.shape)
img[rr,cc,2] = 150


#goals:
rr,cc,val = circle_perimeter_aa(70,50,40)
img[rr,cc,0] = 255
img[rr,cc,1] = 255
rr,cc = rectangle(start=(50,40),end=(90,60))
img[rr,cc,0] = 255
img[rr,cc,1] = 255

rr,cc,val = circle_perimeter_aa(200,300,40)
img[rr,cc,0] = 255
img[rr,cc,1] = 255
rr,cc = rectangle(start=(190,280),end=(210,320))
img[rr,cc,0] = 255
img[rr,cc,1] = 255

rr,cc,val = circle_perimeter_aa(400,300,40)
img[rr,cc,0] = 255
img[rr,cc,1] = 255
rr,cc = rectangle(start=(380,290),end=(420,310))
img[rr,cc,0] = 255
img[rr,cc,1] = 255


#stadium:
rr,cc = rectangle(start=(10,300),end=(130,470))
img[rr,cc,1] = 100
img[rr,cc,0] = 50


#houses:
rr,cc = rectangle(start=(325,10),end=(350,50))
img[rr,cc,0] = 255

rr,cc = rectangle(start=(325,70),end=(360,100))
img[rr,cc,0] = 255

rr,cc = rectangle(start=(325,125),end=(335,150))
img[rr,cc,0] = 255

rr,cc = rectangle(start=(380,20),end=(450,60))
img[rr,cc,0] = 255

rr,cc = rectangle(start=(400,80),end=(430,100))
img[rr,cc,0] = 255


#lale:
rr,cc = rectangle(start=(170,410),end=(290,490))
img[rr,cc,2] = 200
img[rr,cc,1] = 100





ax.axis('off')
ax.set_title('The map')
ax.imshow(img)
plt.show()

```
